# Remove Image Background using Python

Removing background of large dataset by python script

```python
# calculate total execution time

from datetime import datetime
start_time = datetime.now()

print('Start Time: {}'.format(start_time))
```

```python
from rembg import remove
from PIL import Image
import os
```

```python
datadir = 'zz-dataset'
for folder in os.listdir(datadir):
    print(folder + '....')
    path = os.path.join(datadir, folder)
    img_no = 1
    for image in os.listdir(path):
        input_path = os.path.join(datadir, folder, image)
        print('   ------ '+ input_path)
        folder_path = os.path.join('zz-clean-dataset', folder)
        if not os.path.exists(folder_path):
            os.makedirs(folder_path)
        output_path = os.path.join('zz-clean-dataset', folder, folder + '_' + str(img_no) + '.png')
        print('   ------------ '+ output_path)
        img_no += 1

        input = Image.open(input_path)
        output = remove(input)
        output.save(output_path)

```

```python
end_time = datetime.now()
print('End Time: {}'.format(end_time))
print('Duration: {}'.format(end_time - start_time))
```
